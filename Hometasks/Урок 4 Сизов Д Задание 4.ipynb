```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
df=pd.read_csv('creditcard.csv')
data=df.copy()
data.info()
```

```python
class_cnt=data['Class'].value_counts()
print(class_cnt)
```

```python
class_cnt.plot(kind='bar', log=True)
```

```python
plt.hist(data.loc[data['Class']==0,'V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(data.loc[data['Class']==1,'V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.legend()
plt.xlim([-30,-1])
plt.ylim([0,0.13])
```
